```python
#OPERATIONS
a = int(input("enter: "))
b = int(input("enter: "))
print(a+b)
print(a-b)
print(a*b)
print(a/b) #output in float form-quotient
print(a//b) #output was rounded by decimal point(lower value)-quotient
print(a%b) #output is remainder
print(a**b) #power operation
```

```python
#COMPARISON OPERATORS-OUPUTS ARE BOOLEAN
a = 15
b = 27
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
```

```python
#LOGICAL OPERATORS
marks = 10
age = 18
print(age>17 and marks==15)
print(age>17 or marks==15)
print(not(marks==15))
```
